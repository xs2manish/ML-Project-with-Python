```python
def FindDuplicateColumnsInData(DataFrame):
    duplicate_columns={}
    for i in DataFrame.columns:
        duplicate_columns[i]=[]
        for j in DataFrame.drop([i],axis=1).columns:
            #print(j)
            #print(i,j,DataFrame[DataFrame[i]==DataFrame[j]].shape[0]/DataFrame.shape[0])
            if (DataFrame[DataFrame[i]==DataFrame[j]].shape[0]/DataFrame.shape[0])>0.98:
                duplicate_columns[i].append(j)
            else:
                continue
    return duplicate_columns
```

```python
def Nulls(df):
    nulls = pd.DataFrame({'Total Null Values':df.isnull().sum(),'Null %':round(df.isnull().sum()/len(df.index)*100,2)})
    return nulls[nulls['Null %']!=0].sort_values(by='Null %',ascending=False)
```

```python
#Splitting the dataset into the Training set and Test set
X_train_org, X_test_org, y_train, y_test = train_test_split(X, y, random_state = 0, test_size = 0.2)
```

```python
# Scaling the dataset
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train_org)
X_test = scaler.transform(X_test_org)
```

```python
Model_Scoring = pd.DataFrame(columns=['Model','Train Score', 'Test Score','RMSE'])

Model=[]
Train_Score=[]
Test_Score=[]
RMSE=[]
```

###  1. Voting With Hard Classifier Using KNN


```python
knn_clf = KNeighborsClassifier(7)
knn_clf.fit(X_train, y_train)

voting_clf = VotingClassifier(estimators=[('knn', knn_clf)], voting='hard')
voting_clf.fit(X_train, y_train)

y_pred = voting_clf.predict(X_test)

print('Train score For Voting With Hard Classifier Using KNN: {:.4f} %'.format(voting_clf.score(X_train, y_train)*100))
print('Test score For Voting With Hard Classifier Using KNN: {:.4f} %'.format(voting_clf.score(X_test, y_test)*100))
print('RMSE For Voting With Hard Classifier Using KNN: {:.4f}'.format(np.sqrt(mean_squared_error(y_test,y_pred))))

Model.append("Voting With Hard Classifier Using KNN")
Train_Score.append(voting_clf.score(X_train, y_train))
Test_Score.append(voting_clf.score(X_test, y_test))
RMSE.append(np.sqrt(mean_squared_error(y_test,y_pred)))
```

#### As we can see that using Hard Classfier on KNN gives some overfitting, so if we tweak number of neighbours parameter it can be brought to a good accuracy. 

### 2. Voting Wth Soft Classifier Using SoftMax Classifier

```python
softmax_reg = LogisticRegression(multi_class="multinomial",solver="lbfgs", C=10)
softmax_reg.fit(X_train, y_train)

voting_clf = VotingClassifier(estimators=[('softmax_reg', softmax_reg)], voting='soft')
voting_clf.fit(X_train, y_train)

y_pred = voting_clf.predict(X_test)

print('Train score For Voting Wth Soft Classifier Using SoftMax Classifier: {:.4f} %'.format(voting_clf.score(X_train, y_train)*100))
print('Test score For Voting Wth Soft Classifier Using SoftMax Classifier: {:.4f} %'.format(voting_clf.score(X_test, y_test)*100))
print('RMSE For Voting Wth Soft Classifier Using SoftMax Classifier: {:.4f}'.format(np.sqrt(mean_squared_error(y_test,y_pred))))

Model.append("Soft Voting Wth Using SoftMax Classifier")
Train_Score.append(voting_clf.score(X_train, y_train))
Test_Score.append(voting_clf.score(X_test, y_test))
RMSE.append(np.sqrt(mean_squared_error(y_test,y_pred)))
```

#### As we can see that using Soft Classfier on Softmax Classifier gives some overfitting, so if we tweak number parameter C it can be brought to a good accuracy.

```python
def make_meshgrid(x, y, h=.02):
    """Create a mesh of points to plot in

    Parameters
    ----------
    x: data to base x-axis meshgrid on
    y: data to base y-axis meshgrid on
    h: stepsize for meshgrid, optional

    Returns
    -------
    xx, yy : ndarray
    """
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy


def plot_contours(ax, clf, xx, yy, **params):
    """Plot the decision boundaries for a classifier.

    Parameters
    ----------
    ax: matplotlib axes object
    clf: a classifier
    xx: meshgrid ndarray
    yy: meshgrid ndarray
    params: dictionary of params to pass to contourf, optional
    """
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out
```
